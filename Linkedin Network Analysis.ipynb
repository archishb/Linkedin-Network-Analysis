#### Convert the Connected On column to date type 

```python
import datetime
def convert(date):
    return datetime.datetime.strptime(date,"%d %b %Y").strftime("%Y-%m-%d")

df["Connected On"] = df["Connected On"].apply(convert)
```

The apply function applies the function convert on each date value in the Connected On Column.

```python
# sorting the data frame date-wise
df=df.sort_values(by="Connected On")
```

```python
df.head(5)
```

#### Chronological Line Plot of No. of Connections

```python
df_bydate = df.groupby(['Connected On'],as_index=False).count()
```

```python
df_bydate.head(2)
```

```python
connections_line = px.line(df_bydate, x="Connected On", y="First Name", title='My Connections')
connections_line.show()
```

```python
df_bydate['Date']=df_bydate['Connected On']
df_bydate['New_Connections']=df_bydate['First Name']
```

```python
df_bydate.head(2)
```

```python
df_bydate.drop(columns =['Connected On', 'First Name','Last Name','Email Address','Company','Position'], inplace=True ) 
```

```python
df_bydate['Cum'] = df_bydate['New_Connections'].cumsum()
```

```python
df_bydate.head()
```

```python
df_bydate.drop(columns =['New_Connections'], inplace=True ) 
df_bydate.head(3)
```

```python
connections_line = px.line(df_bydate, x="Date", y="Cum", title='My Connections')
connections_line.show()
```

```python
# Company wise histogram
fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="count",  x=df["Company"]))
fig.show()
# Position wise histogram
fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="count",  x=df["Position"]))
fig.show()
```
